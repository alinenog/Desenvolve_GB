```python
def media(lista):
    valor = sum(lista)/len(lista)
    print(valor)
```

```python
media([5,8,6,7,4,6,2])
```

```python
def media(lista):
    valor = sum(lista)/len(lista)
    return(valor, len(lista))
```

```python
#Chamando a função
media([5,8,6,7,4,6,2, 8, 6]) # Valor da média e o tamanho da lista 
```

```python
resultado, n  = media([5,8,6,7,4,6,2, 8, 6])
```

```python
#Resultado da média
resultado
```

```python
#Tamanho da lista
n
```
