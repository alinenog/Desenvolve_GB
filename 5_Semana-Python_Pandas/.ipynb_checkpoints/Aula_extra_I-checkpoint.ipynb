##### 04- Filtrando dados

Relatório de Análise

```python
dados = pd.read_csv('Aluguel.csv', sep=';')
```

```python
dados.head()
```

```python
#Gerando uma lista com os tipos de moveis sem repetição 

list(dados['Tipo'].drop_duplicates())
```

```python
#Lista com imovéis residenciais

residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

```python
residencial
```

```python
dados.head()
```

```python
#Gerar uma siries de True e false quando os itens fores verdadeiros

dados['Tipo'].isin(residencial).head()
```

```python
selecao = dados['Tipo'].isin(residencial)
```

```python
#Gerou um dataframe contendo apenas os itens residenciais

dados_residencial = dados[selecao]
```

```python
dados_residencial
```

```python
#Verificando os tipo de dados

list(dados_residencial['Tipo'].drop_duplicates())
```

```python
#Quantidades dos dados do Df residenciais

dados_residencial.shape[0]
```

```python
#Quantidades dos dados do Df geral

dados.shape[0]
```

```python
#reconstruindo o index

dados_residencial.index = range(dados_residencial.shape[0])
```

```python
dados_residencial
```

Exportando base de dados

```python
dados_residencial.to_csv('aluguel_residencial.csv', sep=';')
```
